# Merging Dataframes

```python
# Merging 2 differnt dataframe into single dataframe
movie_df=pd.merge(left=df1,right=df2,how='inner',left_on='movie_id',right_on='id')

# Dropping one of the common column
movie_df=movie_df.drop('movie_id',axis=1)
movie_df.head(2)
```

```python
# Dropping Home page and tagline which is unique to all the movies
movie_df=movie_df.drop('homepage',axis=1)
movie_df=movie_df.drop('tagline',axis=1)
```

# Handling Json Column

```python
# created a function to convert json to str(string) format
def json_to_string(row):
    json_data = json.loads(row)
    values=[data.get("name") for data in json_data]
    return values[:3] #or else[:1]
```

### Extracting Country name from production_countries

```python
movie_df['countries'] = movie_df["production_countries"].apply(json_to_string)
```

```python
# tmdb_df["genres_extract"] = tmdb_df["genres"].apply(json_to_string)
# tmdb_df["g1"] = tmdb_df["genres_extract"].apply(lambda x:x[0] if len(x)>1 else None)
movie_df["country"] = movie_df['countries'].apply(lambda x:x[0] if len(x)>=1 else None)
```

### Extracting Lead_Actor from cast column

```python
movie_df['cast_def'] = movie_df["cast"].apply(json_to_string)
movie_df["Lead_Actor"] = movie_df['cast_def'].apply(lambda x:x[0] if len(x)>=1 else None)
```

### Extracting Second_Lead_Actor from cast column

```python
movie_df["Second_Lead_Actor"] = movie_df['cast_def'].apply(lambda x:x[1] if len(x)>=2 else None)
```

### Extracting production_house from production_companies column

```python
movie_df['company'] = movie_df["production_companies"].apply(json_to_string)
movie_df["Main_Production_House"] = movie_df['company'].apply(lambda x:x[0] if len(x)>=1 else None)
```

### Extracting Main Generes from genres

```python
movie_df['Genres'] = movie_df["genres"].apply(json_to_string)
# movie_df["Genres"] = movie_df['g1'].apply(lambda x:x[0] if len(x)>=1 else None)
```

### Extracting Director name from crew

```python
movie_df['crew'] = movie_df['crew'].apply(json.loads)
extract_director_names = lambda x: [person['name'] for person in x if person['job'] == 'Director']

movie_df['directors'] = movie_df['crew'].apply(extract_director_names)
movie_df["Director"] = movie_df['directors'].apply(lambda x:x[0] if len(x)>=1 else None)
```

```python
# converting sting to list to merge in later part f the code
movie_df.overview=movie_df.overview.apply(lambda x:x.split())
movie_df.Lead_Actor=movie_df.Lead_Actor.apply(lambda x:x.split())
movie_df.Second_Lead_Actor=movie_df.Second_Lead_Actor.apply(lambda x:x.split())
```

```python
#removing the space eg firstname and lastname
movie_df.Genres=movie_df.Genres.apply(lambda x:[i.replace(" ","") for i in x])
movie_df.keywords=movie_df.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movie_df.Lead_Actor=movie_df.Lead_Actor.apply(lambda x:[i.replace(" ","") for i in x])
movie_df.Second_Lead_Actor=movie_df.Second_Lead_Actor.apply(lambda x:[i.replace(" ","") for i in x])
movie_df.directors=movie_df.directors.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_df['Tags']=movie_df.overview+movie_df.Genres+movie_df.keywords+movie_df.Lead_Actor+movie_df.Second_Lead_Actor+movie_df.directors
```

```python
#Selecting few columns for recomendation system
new_movie_df=movie_df[['title_x','id','Tags']]
```

```python
#Convert list to string
new_movie_df.Tags=new_movie_df.Tags.apply(lambda x:' '.join(x))
```

```python
new_movie_df
```

```python
#convering to lowercase
new_movie_df.Tags=new_movie_df.Tags.apply(lambda x:x.lower())
```

```python
new_movie_df
```

```python
#some of the words are same like accept and accepts etc we have to make that to one word.
# Stemming Techinque to make similar words to one word
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_movie_df.Tags=new_movie_df.Tags.apply(stem)
```

```python
# Removing stop words eg to and a etc and giving max words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movie_df.Tags).toarray()
```

```python
vectors[0]
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
# to find the distance of movie with other movies.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simalarity=cosine_similarity(vectors)
```

```python
# 1st movie with same movie score of the distance will be 1
simalarity[0]
```

```python
# Recomendation System
def recommand(movie):
    movie_index=new_movie_df[new_movie_df.title_x==movie].index[0] #finding index of movie
    distances=simalarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] #sorting top 5 movie with holding the index value 
    
    for i in movie_list: #removing the movie in side tuple holding index value
        print(new_movie_df.iloc[i[0]].title_x)
        print()
```

```python
recommand('Titanic')
```
